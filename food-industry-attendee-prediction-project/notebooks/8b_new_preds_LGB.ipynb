```python
## load model
model = pickle.load(open(r'.\models\lightGBMmodel.sav', 'rb'))
data = pd.read_pickle(r'.\data\output\dataForPreds.pkl')

def predict_lgb(data):
    ids = data['10 Digit Card Number']
    data = data.drop(['10 Digit Card Number', 'show', 'latitude', 'longitude', 'count_per_comp_website'], axis=1)
    y_pred = model.predict(data)
    y_pred = pd.DataFrame(y_pred)
    y_pred['10 Digit Card Number'] = ids
    return y_pred

y_pred = predict_lgb(data)
```

```python
y_pred.to_csv(r'.\data\output\lightGBM_predictions.csv')
```
